### Spliting the data

```python
X=df.drop(columns='target',axis=1)
Y=df['target']
```

```python
X.head()
```

```python
Y.head()
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

```python
X_train
```

# Scaling

```python
from sklearn.preprocessing import Normalizer
transformer = Normalizer().fit(X)
X = transformer.transform(X)
```

```python
X_train_scaled, X_test_scaled, Y_train_scaled, Y_test_scaled= train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train_scaled.shape,X_test_scaled.shape)
```

### Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
model.fit(X_train,Y_train)
```

### Logistic Regression Model Evaluation

```python
lg_X_train_prediction=model.predict(X_train)
lg_training_data_accuracy=accuracy_score(lg_X_train_prediction,Y_train)
lg_training_data_accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(lg_X_train_prediction,Y_train)
print(cm)
```

```python
ax=plt.axes()
sns.heatmap(cm,annot=True,fmt='d',cmap='Blues',ax=ax,annot_kws={'size':30})
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(lg_X_train_prediction,Y_train))
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,X_train,Y_train,cv=5)
scores
```

```python
scores.mean()
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
test_data_accuracy
```

### Prediction

```python
input_data=(41,0,1,130,204,0,0,172,0,1.4,2,0,2)

# Conveting to numpy array
input_data_as_numpy_array=np.asarray(input_data)

# reshape the array
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
```
